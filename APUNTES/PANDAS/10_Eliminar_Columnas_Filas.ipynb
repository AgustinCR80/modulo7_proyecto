# Método .drop()

# Sintaxis:

# df.drop(labels, axis = 0, inplace = False)
'''
Donde:
- df: Nombre del DataFrame que vamos a utilizar.
- labels: nombres de las columnas a eliminar, si es más de una lo hacemos como lista ['__', '__']
- axis: Especifica si queremos eliminar filas (0) o columnas (1)
- inplace (opcional): Indica si se va a realizar la eliminación sobre el DataFrame original (True), es decir, lo sobrescribe, ó si va a crear un DataFrame nuevo (False)
'''

```python
# Importamos las librerias y el archivo csv:
import pandas as pd
pd.set_option('display.max_columns', None)  # Para que aparezcan todas las columnas

path = 'C:/Users/Agus_Ana_María_Raúl/Agustin/DAM/MASTER/MÓDULO_7_PYTHON_FOR_DATA/2_PANDA/DatosPandasUniversidades.csv'
df_universidades = pd.read_csv(path, index_col = 2)  # Asignamos la tercera columna del csv como índice
df_universidades.head()
```

# Eliminar columna 'sequence'

```python
# La eliminamos del dataframe original:
df_universidades.drop('sequence', axis = 1, inplace = True)
```

```python
# Comprobamos el dataframe original:
df_universidades.head()
```

# Eliminar la fila 'Massachusetts Institute of Technology (MIT)'

```python
# No la eliminamos del dataframe original:
df_universidades.drop('Massachusetts Institute of Technology (MIT)', axis = 0, inplace = False)
```

```python
# Comprobamos que no la hemos eliminado del datafrme original:
df_universidades.head()
```

```python
# Guardamos el dataframe sin la fila 'Massachusetts Institute of Technology (MIT)' en una variable:
df_universidades_sin_Massachusetts = df_universidades.drop('Massachusetts Institute of Technology (MIT)', axis = 0, inplace = False)
```

```python
# Comprobamos la variable:
df_universidades_sin_Massachusetts.head()
```
