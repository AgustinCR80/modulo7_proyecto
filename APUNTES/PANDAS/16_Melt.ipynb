# Melt()

# Nos permite transformar un DataFrame con muchas columnas en otro con menos, donde una o más son variables de identificación y otras se fusionan.

# Sintaxis:

pd.melt(df, id_vars = None, value_vars = None, var_name = None, value_name = None)

· df: DataFrame que vamos a usar.

· id_vars: Columnas que serán las variables identificadoras. Estas no van a cambiar al generar el nuevo DataFrame.

· value_vars: Columnas que se van a fundir o fusionarse.

· var_name: Nombre de la nueva o nuevas columnas generadas por la unión de value_vars.

· value_name: Nombre de la nueva o nuevas columnas con los valores de los registros de value_vars.

# Fusionar las columnas, ’numero_puertas’, ‘camara_trasera’ y sensor_aparcamiento’, usaremos como índice el modelo del coche.

```python
# Importamos las librerias y el archivo csv:
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)  # Para que aparezcan todas las columnas
```

```python
# Importar archivos:

path1 = 'C:/Users/Agus_Ana_María_Raúl/Agustin/DAM/MASTER/MÓDULO_7_PYTHON_FOR_DATA/2_PANDA/datos_ford_2024.csv'
coches_2024 = pd.read_csv(path1, index_col = 0)

path2 = 'C:/Users/Agus_Ana_María_Raúl/Agustin/DAM/MASTER/MÓDULO_7_PYTHON_FOR_DATA/2_PANDA/datos_ford_caracteristicas.csv'
coches_caracteristicas = pd.read_csv(path2, index_col = 0)
```

```python
# Realizamos un merge() para combinar los dos dataframe (columna VIN común a los dos):
coches_merge = coches_2024.merge(coches_caracteristicas, how = 'inner', on = 'VIN')
coches_merge.head()
```

```python
# Convertimos el merge en 5 filas para que no sea tan largo
coches_merge = coches_merge.head(5)
coches_merge
```

```python
# pd.melt(df, id_vars = None, value_vars = None, var_name = None, value_name = None)

coches_melt = pd.melt(coches_merge, id_vars = 'modelo_coche', value_vars = ['numero_puertas', 'camara_trasera', 'sensor_aparcamiento'], var_name = 'Características', value_name = 'Valor')
coches_melt

# Los 1 y 0 de la columna valor que corresponden a camara_trasera y sensor_aparcamiento son los True y False de esas columnas en coches_merge
```
