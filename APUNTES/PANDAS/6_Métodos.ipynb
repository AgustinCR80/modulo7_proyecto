```python
# Los dos métodos anteriores, isnull() y isna(), por si mismos nos dan poca información. 
# Es mucho más cómodo usar el método .sum() que va a ir sumando los True (que sí sean nulos) que encuentre.
# Devuelve el nombre de la columna y los nulos que hay en su interior.

df_universidades.isnull().sum()
```

```python
# Contar los nulos en una columna, 'country'
df_universidades['country'].isnull().sum()
```

# df.describe()

```python
# Indica los principales datos estadísticos del DataFrame. Por defecto nos los calcula de las variables numéricas, no objects, str...
# Contaje total de registros(filas), mediana, desviación estandar, valor mínimo, los tres cuartiles y el valor máximo

df_universidades.describe()
```

```python
# Para que sea más amigable ver los resultados transponemos las filas por columnas u viceversa.

df_universidades.describe().T
```

```python
# Podemos utilizar .describe() para una columna 'rank'

df_universidades['rank'].describe()
```

```python
# Podemos obtener los datos estadísticos por separado de la columna 'rank'

print(f'Count: {df_universidades['rank'].count()}')  # Cuenta el número de valores no nulos de la columna: 105
print(f'Mean: {df_universidades['rank'].mean()}')    # Calcula la media de los valores de la columna: 52.8
print(f'Min: {df_universidades['rank'].min()}')    # Muestra el valor mínimo de nuestra columna: 1
print(f'Q25: {df_universidades['rank'].quantile(0.25)}')  # Calcula el 1º cuartil (Q1) o percentil 25 de los valores de la columna: 26.0
```

```python
# Utilizamos el método con un dato categórico, object:

df_universidades.describe(include = 'object').T

# Salida: nombre de la columna, contaje total de registros, valores únicos, el que más se repite (la moda) y cuantas veces se repite
```
