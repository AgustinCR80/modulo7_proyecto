## Convertir columnas numéricas a categóricas

### Convertir 'y' a columna categórica

```python
bank_concat['y'] = bank_concat['y'].apply(lambda x: 'yes' if x == 1.0 else 'no')
bank_concat['y'].value_counts(dropna=False)
```

### Convertir 'default', 'housing', 'loan' a columnas categórica

```python
def cambiar_valor_a_str(valor):
    """_summary_ Cambia el valor a 'no' si el valor inicial es 0.0, a 'yes', si es 1.0, Nan lo mantiene
    Args:
        valor (float or null): 0.0 - 1.0 - Nan 

    Returns:
        _type_: Devuelve el cambio según la condición
    """
    if valor == 0.0:
        return 'no'
    
    elif valor == 1.0:
        return 'yes'
    
    else:
        return valor
```

```python
columnas = ['default', 'housing', 'loan']

for col in columnas:
    bank_concat[col] = bank_concat[col].apply(cambiar_valor_a_str)

# Utilizando applymap podemos llamar a la función conjuntamente con otras columnas.
# bank[['default', 'housing', 'loan']] = bank[['default', 'housing', 'loan']].applymap(cambiar_valor_a_str)

bank_concat.head(5)
```

## Columnas numéricas

### Scatterplot

```python
# Quitar la columna 'Income' de las columnas numéricas para no realizar el gráfico por si misma dentro del bucle for
# Utilizo 'Income' en vez de 'index' porque se aprecian mejor los gráficos
columna = 'Income' 
cols_num = bank_concat.select_dtypes(include='number').drop(columns=[columna]).columns

# Generar gráficos de dispersión:
for col in cols_num:

    # Inicializar una figura
    plt.figure()

    # Customizar gráfico
    sns.scatterplot(x=bank_concat[col], y=bank_concat[columna], hue=bank_concat['y'], alpha=0.3)
    plt.title(f'{col}-{columna}')
    plt.xlabel(f'{col}')
    plt.ylabel(f'{columna}')
    plt.grid(alpha= 0.4)  # Mostar una cuadrícula e intensidad de las líneas
    plt.xticks(rotation=0) # Rotación etiquetas

    # Mostrar la figura
    plt.show()
```

##### La mayoría de los clientes que contrataron el producto 'y' están comprendidos aprox. entre el índice 37000 y 41000, se había contactado con ellos antes de la campaña al menos dos veces y cuanto más larga es la última llamada 'duration' es más probable que el cliente contrate el producto. Existe una correlación negativa en la tasa de interés del euribor a tres meses, 'nr.employed' (número de empleados, gente trabajando) y 'emp.var.rate' (tasa de variación del empleo),  lo que significa que cuando el euribor baja, en nuestro caso por debajo del 1%, y la economía va peor porque hay menos gente trabajando y la demanda de empleo es baja, los clientes contratan depósitos a plazo porque lo ven como un producto seguro. Los clientes comprendidos en esos índices pueden haber sido captados en un momento de la campaña más efectiva o por un contexto económico propenso a la contratación del plazo.
